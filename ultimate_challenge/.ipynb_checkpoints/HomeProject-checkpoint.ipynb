## Part 1 â€” Exploratory data analysis
The attached logins.json file contains (simulated) timestamps of user logins in a particular geographic location. Aggregate these login counts based on 15minute time intervals, and visualize and describe the resulting time series of login counts in ways that best characterize the underlying patterns of the demand. Please report/illustrate important features of the demand, such as daily cycles. If there are data quality issues, please report them.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

````python
# Load the dataset
logins_df = pd.read_json('logins.json')

# Convert to datetime
logins_df['login_time'] = pd.to_datetime(logins_df['login_time'])

# Set the login_time column as the index
logins_df.set_index('login_time', inplace=True)

# Resample to 15-minute intervals and count occurrences
login_counts = logins_df.resample('15T').size()

# Convert Series to DataFrame for clarity and better handling
login_counts = login_counts.to_frame(name='login_count')
````

````python
# Plot the login counts over time
plt.figure(figsize=(10,6))
plt.plot(login_counts.index, login_counts['login_count'])
plt.title('Login Counts Over Time (15-Minute Intervals)')
plt.xlabel('Time')
plt.ylabel('Login Count')
plt.show()
````

Based on the pattern, it appears that there is significant fluctuation in login counts over time, with some notable spikes in activity. 

### 1. Daily Cycle Patterns

It could be useful to analyze the daily cycle by aggregating the data into hourly or daily intervals to observe if there is a recurring pattern in user activity over the day.

````python
# Aggregating by hour to see daily patterns
login_counts['hour'] = login_counts.index.hour
hourly_pattern = login_counts.groupby('hour')['login_count'].mean()

# Plotting the daily pattern
plt.figure(figsize=(10,6))
plt.plot(hourly_pattern.index, hourly_pattern.values)
plt.title('Average Login Counts by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Average Login Count')
plt.show()
````

### 2. Weekly Cycle Patterns

````python
# Aggregating by day of the week
login_counts['day_of_week'] = login_counts.index.dayofweek
weekly_pattern = login_counts.groupby('day_of_week')['login_count'].mean()

# Plotting the weekly pattern
plt.figure(figsize=(10,6))
plt.plot(weekly_pattern.index, weekly_pattern.values)
plt.title('Average Login Counts by Day of the Week')
plt.xlabel('Day of the Week (0 = Monday)')
plt.ylabel('Average Login Count')
plt.show()
````

### 3. Smoothing the Data

The raw time-series data can be noisy, as shown by the sharp spikes.I applied smoothing techniques like a rolling mean to better observe the trends over time.

````python
# Applying a rolling mean to smooth out short-term fluctuations
login_counts['rolling_mean'] = login_counts['login_count'].rolling(window=96).mean()  # 96 points for 24 hours (15 min intervals)

# Plotting the smoothed data
plt.figure(figsize=(10,6))
plt.plot(login_counts.index, login_counts['rolling_mean'], color='orange')
plt.title('Smoothed Login Counts Over Time (Rolling Mean)')
plt.xlabel('Time')
plt.ylabel('Login Count (Smoothed)')
plt.show()
````

**Key Observations from the Smoothed Data:**

Cyclical Behavior: There is a clear cyclical pattern with regular rises and falls in the login counts, possibly indicating daily or weekly cycles.
Increasing Trend: As time progresses, the peaks of the cycles appear to be increasing, indicating growing activity or demand over time.

### Seasonal decomposition

````python
from statsmodels.tsa.seasonal import seasonal_decompose

# Seasonal decomposition of the login counts
decomposition = seasonal_decompose(login_counts['login_count'], model='additive', period=96)  # 96 intervals = 1 day (15-min intervals)

# Plot the decomposed components
decomposition.plot()
plt.show()
````

**Key Insights:**

The trend indicates a growing number of logins over time, with regular cycles of high activity followed by drops, suggesting that the platform is gaining more usage over time.

The seasonality shows a clear pattern that recurs within shorter intervals, likely daily or hourly patterns of user logins.

The residual shows that while the trend and seasonality explain much of the data, there are occasional outliers or noise, particularly visible towards the end of the series.

## **Part 2 â€” Experiment and Metrics Design**

### **Scenario Recap:**
Gotham and Metropolis are neighboring cities with different activity patterns:
- **On weekdays**: Gotham is most active at night, and Metropolis during the day.
- **On weekends**: Both cities have reasonable activity.
  
However, thereâ€™s a **toll bridge** between the two cities, and the toll discourages drivers from serving both cities. The idea is to **reimburse the toll costs** to encourage drivers to work in both cities. We need to design an experiment to test if this reimbursement will increase driver activity in both Gotham and Metropolis.

### 1) **Key Measure of Success**

The goal of the experiment is to determine if reimbursing toll costs encourages drivers to serve **both cities**. The best metric should reflect the **increase in driver activity** across both Gotham and Metropolis.

#### **Key Metric:**
- **Number of trips between Gotham and Metropolis** (cross-city trips).
  The purpose of reimbursing the toll is to motivate drivers to cross the toll bridge and serve both cities. By tracking the number of trips made between the two cities, we can directly measure if the reimbursement is effective in achieving this goal.

---

### 2) **Experiment Design**


#### **a) How will you implement the experiment?**

1. **Split the drivers into two groups**:
   - **Control group**: These drivers do **not receive toll reimbursement**. They operate as they normally do.
   - **Test group**: These drivers **receive full toll reimbursement** every time they cross the bridge between Gotham and Metropolis.

2. **Duration of the experiment**: 
   - The experiment should run for at least **one month**. This period will give enough time to observe changes in driver behavior and see if the reimbursement has a consistent impact over time.

3. **Data collection**:
   - **Before the experiment**: Gather baseline data on how many drivers are making cross-city trips and how frequently they are doing it.
   - **During the experiment**: Track the number of trips between Gotham and Metropolis for both the control and test groups.

4. **What to measure**:
   - **Number of cross-city trips**: Measure how many trips each group is making between Gotham and Metropolis.
   - **Percentage of drivers** who are working in both cities: Measure the percentage of drivers in each group who are serving both cities, not just one.

#### **b) Statistical test(s) to verify the significance of the observation**

1. **Statistical Test**:
   - Use a **t-test** to compare the **average number of trips between Gotham and Metropolis** for the control group and the test group. The t-test will help determine if the difference in the number of trips between the two groups is statistically significant.
   
   - If the data does not follow a normal distribution, i will use a non-parametric test like the **Mann-Whitney U test** to compare the groups.

```python
from scipy import stats

# Example t-test to compare the two groups
t_stat, p_value = stats.ttest_ind(control_group_trips, test_group_trips)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

2. **Interpretation of the results**:
   - If the **p-value** is **less than 0.05**, it means there is a **statistically significant difference** between the control group and the test group. This would mean that the toll reimbursement had a real impact on increasing the number of cross-city trips.
   - If the **p-value** is **greater than 0.05**, it means there is **no significant difference**, and the toll reimbursement did not have a noticeable effect.

#### **c) Interpreting the results and providing recommendations**

After analyzing the results, here are the two possible outcomes:

1. **If the toll reimbursement significantly increases cross-city trips** (p-value < 0.05):
   - **Conclusion**: The experiment shows that toll reimbursement is effective in encouraging drivers to work in both cities.
   - **Recommendation**: Implement the toll reimbursement policy on a permanent basis. Since it encourages more drivers to cross the toll bridge and serve both cities, it helps balance driver availability in both Gotham and Metropolis.

2. **If the toll reimbursement does not significantly increase cross-city trips** (p-value > 0.05):
   - **Conclusion**: The reimbursement of toll costs did not significantly change driver behavior.
   - **Recommendation**: Do not implement the reimbursement policy. Instead, explore other ways to motivate drivers to work in both cities, such as offering bonuses for cross-city trips or introducing a different incentive structure.

---

### **Summary of Experiment Design**

1. **Key Measure of Success**: 
   - The number of trips between Gotham and Metropolis (cross-city trips).

2. **Experiment Design**:
   - Randomly divide drivers into two groups (control and test groups).
   - Test group receives toll reimbursement; control group does not.
   - Run the experiment for one month and track the number of cross-city trips.

3. **Statistical Test**:
   - Use a **t-test** to compare the average number of trips in the control and test groups.
   - A significant result (p-value < 0.05) means the reimbursement works.

4. **Recommendations**:
   - If the test shows a significant increase in cross-city trips, implement the policy.
   - If thereâ€™s no significant increase, consider other strategies to encourage drivers to work in both cities.

Retention Calculation:

We define retention as a user who was active (i.e., took a trip) in the last 30 days of the dataset. To implement this:

We first identified the latest date in the dataset, and then created a cutoff date by subtracting 30 days from it.
We created a new column, retained, which marks whether a user is retained (1) or not (0).

````python
# Define the cutoff date for retention
last_date = data['last_trip_date'].max()
cutoff_date = last_date - pd.Timedelta(days=30)

# Creating the 'retained' column
data['retained'] = (data['last_trip_date'] >= cutoff_date).astype(int)

# Checking the fraction of retained users
retained_fraction = data['retained'].mean()
print(f"Fraction of retained users: {retained_fraction:.2f}")
````
